# Uranium Ore Processing

Miner + Sulfuric Acid -> Uranium Ore

Uranium Ore + Centrifuge -> U235 + U238

10 Iron Plate + 1 U235 + 19 U238 -> 10 Uranium Fuel Cell



```python
from math import gcd
```

To find the least common multiple:

(a*b)/gcd(a,b)

```python
a = 10
b = 19

(a*b)/gcd(a,b)
```

- 10 sulfuric acid per 10 ore mined
- 1 barrel = 50 fluid

```python
((1*50)/10)*10
```

```python
from graphviz import Digraph

f = Digraph('uranium_ore_processing')
f.attr(rankdir='LR', size='8,5')

f.edge('Uranium Ore', 'U235', label='Centrifuge')
f.edge('Uranium Ore', 'U238', label='Centrifuge')

f
```

```python
from graphviz import Digraph

f = Digraph('uranium_fuel_cell')
f.attr(rankdir='LR', size='8,5')

f.edge('Iron Plate', 'Uranium Fuel Cell', label='Assembling Machine')
f.edge('U235', 'Uranium Fuel Cell', label='Assembling Machine')
f.edge('U238', 'Uranium Fuel Cell', label='Assembling Machine')

f
```
